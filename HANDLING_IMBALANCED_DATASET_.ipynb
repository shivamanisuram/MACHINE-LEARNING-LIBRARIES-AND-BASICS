```python
#distribution of 2 classes
credit_card_data['Class'].value_counts()
```

```python
#0----->represents legit transactions
#1------>represnts fradulent transaction
```

```python
legit=credit_card_data[credit_card_data.Class==0]
fradulent=credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
```

```python
print(fradulent.shape)
```

UNDER SAMPLING:build a sample dataset containing legid and fradulent transactions

no of fradulent transactions---->163

```python
legit_sample=legit.sample(n=163)
```

```python
print(legit_sample.shape)
```

```python
#conactenate two dataframes
```

```python
new_dataset=pd.concat([legit_sample,fradulent],axis=0)
```
