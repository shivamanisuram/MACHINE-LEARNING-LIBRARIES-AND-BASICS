```python
#house price dataset
# Instead of using load_boston, we will use fetch_california_housing as an alternative
from sklearn.datasets import fetch_california_housing # Import fetch_california_housing
housing = fetch_california_housing() # Use fetch_california_housing to load the data
house = pd.DataFrame(housing.data, columns=housing.feature_names) # Create DataFrame
house['price'] = housing.target # Assuming 'price' is the target variable
```

CORRELATION:
1.POSTIVE CORRELATION
2.NEGATIVE CORRELATION

```python
# Convert the housing data to a Pandas DataFrame first
housing_df = pd.DataFrame(housing.data, columns=housing.feature_names)
housing_df['price'] = housing.target # Add the target variable to the DataFrame

# Now you can calculate the correlation using the DataFrame
correlation = housing_df.corr()
```

```python
#HEAT MAP
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```
