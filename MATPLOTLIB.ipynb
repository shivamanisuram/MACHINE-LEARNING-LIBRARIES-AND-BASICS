MATPLOTLIB:
USEFUL FOR MAKING PLOTS

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

plotting data


```python
#sin wave
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.show()
```

```python
#adding title to x-axix & y-axis labels:
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('sine wave')
plt.show()
```

```python
#parabola
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('parabola')
plt.show()
```

```python
plt.plot(x,y,'r+')
```

```python
plt.plot(x,y,'b.')
```

```python
x=np.linspace(-5,5,100)
plt.plot(x,np.sin(x),'r--')
plt.plot(x,np.cos(x),'yx')
plt.show()
```

BAR PLOT

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
langs=['c','c++','java','python','php']
students=[23,17,35,29,12]
ax.bar(langs,students)
plt.xlabel('languages')
plt.ylabel('number of students')
plt.show
```

PIE CHART

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
langs=['c','c++','java','python','php']
students=[23,17,35,29,12]
ax.pie(students,labels=langs,autopct='%1.2f%%')
plt.show()
```

scatter plot


```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

3D SCATTER PLOT


```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Greens')
plt.show()
```
